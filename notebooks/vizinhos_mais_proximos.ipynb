```python
x = pontos['geometry'].x
y = pontos['geometry'].y
z = pontos['Elevation']
```

total_bounds atributo Ã© usado para pegar os mÃ¡ximos e mÃ­nimos das coordenadas de x e y na seguinte ordem:<br> (minx, miny, maxx, maxy)

```python
pontos.total_bounds
```

```python
minx, miny, maxx, maxy = pontos.total_bounds
print(f"mÃ­nima coordenada de x: {minx}\nmÃ¡xima coordenada de x: {maxx}\nmÃ­nima coordenada de y: {miny}\nmÃ¡xima coordenada de y: {maxy}")
```

```python
grid_de_x = np.arange(minx - 10,maxx + 10, 10, dtype='float64')
grid_de_x
```

```python
grid_de_y = np.arange(miny - 10, maxy + 10, 10, dtype='float64')
grid_de_y
```

Antes de criar o grid utilizando meshgrid da numpy, vou fazer um exemplo pequeno para melhor compreensÃ£o:

```python
arrayx= np.arange(0,5,1)
arrayy = np.arange(0,10,1)
AX, AY = np.meshgrid(arrayx, arrayy)
AX, AY
```

```python
xi, yi = np.meshgrid(grid_de_x, grid_de_y)
```

```python
from scipy.interpolate import NearestNDInterpolator

interpolador = NearestNDInterpolator(list(zip(x, y)), z)
```

```python
valores_interpolados = interpolador(xi, yi)
```

```python
list(zip(x,y))
```

```python
valores_interpolados
```

```python
fig,ax = plt.subplots(figsize = (8,10))

cax = ax.imshow(valores_interpolados, extent=(minx-5, maxx+5, miny-5, maxy+5), origin="lower", cmap="YlOrRd")

pontos.plot(color="blue", ax=ax, marker="+", label="Amostras")

cbar=plt.colorbar(cax, fraction = 0.08)

ax.legend()
plt.title("InterpolaÃ§Ã£o vizinho mais prÃ³ximo")

plt.show()

```
